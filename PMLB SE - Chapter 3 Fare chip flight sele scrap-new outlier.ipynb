### Applied city, date and fare to this model

```python
Date=[]
City=[]
Fare=[]
```

```python
from datetime import date, timedelta
from time import sleep
```

```python
# replace this with the path of where you downloaded chromedriver
chromedriver_path = "C:/Users/DUY/Downloads/chromedriver.exe"
browser = webdriver.Chrome(chromedriver_path)
```

```python
start_sat = '2020-05-09'
end_sat = '2020-05-20'

start_sat_date = datetime.strptime(start_sat, '%Y-%m-%d')
end_sat_date = datetime.strptime(end_sat, '%Y-%m-%d')

fare_dict = {}

for i in range(26):
    print('i: ',i)
    sat_start = str(start_sat_date).split()[0]
    sat_end = str(end_sat_date).split()[0] 
    print(sat_start)
    print(sat_end)
    
    fare_dict.update({sat_start: {}})

    #sats = 'https://www.google.com/flights/?f=0#f=0&flt=/m/02_286.r/m/02j9z.2018-12-01*r/m/02j9z./m/02_286.2018-12-08;c:USD;e:1;s:0*1;sd:1;t:e'
    
    sats = "https://www.google.com/flights/?f=0#f=0&flt=/m/02_286.r/m/02j9z." + \
    sat_start + "*r/m/02j9z./m/02_286." + \
    sat_end + ";c:USD;e:1;s:0*1;sd:1;t:e"
    
    sleep(np.random.randint(3,7))
    browser.get(sats)
    print(browser.title)
    
    sleep(np.random.randint(3,7)) # Must have time for page loading before import all data to BeautifulSoup
    soup = BeautifulSoup(browser.page_source, "html5lib")
    cards = soup.select('div[class*=tsAU4e]')
    #cards[0]
    j=0
    for card in cards:
        j+=1
        try:
            sat_start = str(start_sat_date).split()[0]
            sat_end = str(end_sat_date).split()[0]  
            fare_dict.update({sat_start: {}})

            city= card.select_one('h3').text
            print(city)
            fare=card.select_one('div.Q70fcd.sSHqwe').get_text()
            print(fare)
            
            Date.append(sat_start)
            City.append(city)
            Fare.append(fare)
            
            fare_dict[sat_start] = {**fare_dict[sat_start], **{city: fare}}

        except:
            pass

    start_sat_date = start_sat_date + timedelta(days=7)
    end_sat_date = end_sat_date + timedelta(days=7)
```

```python
df=pd.DataFrame(list(zip(Date,City,Fare)),columns=['Date','City','Fare'])
df
```

## Begin Data analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv('C:\\Users\\DUY\\Data science\\ML Blue print Packt\FareFlight2.csv')
df.head()
```

```python
df.dtypes
```

```python
df['City'].unique()
```

```python
df['Fare'][1]
```

```python
df['Fare'] = df['Fare'].map(lambda x: str(x).replace('\xa0₫',''))
df['Fare'] = df['Fare'].map(lambda x: str(x).replace('.','')).astype('int')
df
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.head(3)
```

```python
fig,ax = plt.subplots(figsize=(10,6))
df.groupby('Date').agg({'Fare':'mean'}).plot(ax=ax)
df.groupby('Date').agg({'Fare':'std'}).plot(secondary_y=True,ax=ax)
```

```python
table=pd.pivot_table(df,values='Fare',index=['Date'],columns='City',aggfunc={'Fare':'mean'})
table.head(5)
```

```python
table.columns
```

```python
fig,ax = plt.subplots(figsize=(10,6))
table.plot(y=['Am-xtéc-đam','Pa ri','Luân Đôn'],marker='o', use_index=True,ax=ax)
ax.set_xticks(table.index)
ax.set_xticklabels(table.index, rotation=90);
plt.tight_layout()
```

```python
for name in table.columns:
    #fig,ax = plt.subplots(figsize=(10,3))
    table.plot(y=name,marker='o', use_index=True)
    ax.set_xticks(table.index)
    ax.set_xticklabels(table.index, rotation=90);

plt.tight_layout()
```

```python
df_group_date=df.groupby('Date').agg({'Fare':['mean','min','max']}).reset_index()
df_group_date
```

```python
df_group_city=df.groupby('City').agg({'Fare':'mean'}).reset_index()
df_group_city.sort_values(by='Fare',ascending=False) # Giá rẻ nhất theo quốc gia
```

```python
city='Luân Đôn'
newdf=df[df['City']==city]
fig,ax = plt.subplots(figsize=(10,6))
plt.scatter('Date', 'Fare', color='blue', s=50,data=newdf)
ax.set_xticks(newdf['Date'])
ax.set_xticklabels(newdf['Date'], rotation=-70);
```
